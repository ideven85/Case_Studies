```python
import numpy as np
```

```python
v = np.array([1, 3])
w = np.array([5, 2])
np.cos(30)
```

```python
v_norm = np.linalg.norm(v)
v_norm
```

```python
def cosine_similarity(a, b):
    num = a @ b
    distance_a = np.sqrt(np.square(a[0]) + np.square(a[1]))
    distance_b = np.sqrt(np.square(b[0]) + np.square(b[1]))
    return num / (distance_a * distance_b)


cosine_similarity(v, w)
```

```python
from scipy.spatial.distance import hamming, cosine

print(hamming(v, w))
print(cosine(v, w))
```
