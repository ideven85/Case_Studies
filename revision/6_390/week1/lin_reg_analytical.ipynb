```python
pollution = pd.DataFrame(
    {
        "Temperature": [90, 25, 35],
        "Population": [45, 32, 100],
        "Pollution": [7.2, 9.5, 8.4],
    },
    index=["Chicago", "New York", "Boston"],
)
```

```python
# centering the data
pollution["Temperature"] = pollution["Temperature"] - pollution["Temperature"].mean()
```
